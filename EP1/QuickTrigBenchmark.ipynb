```python
import QuickTrig
import math
import matplotlib.pyplot as plt
%matplotlib inline

x = []
ysin = []
ycos = []
ytan = []
ysec = []
ycsc = []
ycot = []


for n in range(-10000, 10000):
    x.append((math.pi*2/10000.0)*n)
    ycos.append(QuickTrig.cos((math.pi*2/10000.0)*n))
    ysin.append(QuickTrig.sin((math.pi*2/10000)*n))
    ytan.append(QuickTrig.tan((math.pi*2/10000)*n))
    ysec.append(QuickTrig.sec((math.pi*2/10000)*n))
    ycsc.append(QuickTrig.csc((math.pi*2/10000)*n))
    ycot.append(QuickTrig.cot((math.pi*2/10000)*n))
    
```

Gr치fico para seno:

```python
plt.plot(x, ysin)
plt.show()
```

Gr치fico para cosseno:

```python
plt.plot(x, ycos)
plt.show()
```

Gr치fico para tangente:

```python
plt.axis([-4, 4, -10, 10])
plt.plot(x, ytan)
plt.show()
```

Gr치fico para secante:

```python
plt.axis([-4, 4, -10, 10])
plt.plot(x, ysec)
plt.show()
```

Gr치fico para cossecante:

```python
plt.axis([-4, 4, -10, 10])
plt.plot(x, ycsc)
plt.show()
```

Gr치fico para cotangente:

```python
plt.axis([-4, 4, -10, 10])
plt.plot(x, ycot)
plt.show()
```
